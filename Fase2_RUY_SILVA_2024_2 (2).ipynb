# Questão 1: Pacman

**Descrição:**
\
O Pacman é um jogo arcade criado em 1980, o principal objetivo do jogo é coletar o máximo de frutinhas sem encostar nos fantasmas (que causam o fim do jogo). Aqui seguiremos uma premissa parecida na qual queremos encontrar o número máximo de frutinhas que podem ser coletadas num determinado tabuleiro, mas com fantasmas camaradas que ao invés de finalizarem o jogo apenas roubam nossas frutas.  



**Tarefa:**
\
Você deverá escrever um programa que retorne o número máximo de frutinhas que podem ser capturadas em um dado tabuleiro seguindo as seguintes regras:

* O Pacman sempre irá começar no lado superior esquerdo do tabuleiro.
* O Pacman pode terminar o jogo a qualquer momento levando consigo as frutas que carrega no momento.
* Sempre que o Pacman encostar em um fantasma ele perde todas as frutas que está carregando
* Ele percorrerá a primeira linha da esquerda para a direita até chegar o fim da linha.
* Ao chegar ao final de uma linha, ele irá descer uma posição e percorrer toda a linha, dessa vez no sentido oposto da linha anterior.
* A última regra será repetida até que o tabuleiro inteiro tenha sido percorrido.

**Entrada**:
\
Seu programa irá receber como entrada o tamanho do tabuleiro e as strings que irão compor o mesmo, seguindo as seguintes regras:

* A primeira informação de entrada será um número inteiro N que corresponde ao tamanho do tabuleiro (Que sempre será quadrado)
* As N seguinte entradas terão N caracteres que podem ser:
  * "." Um espaço vazio do tabuleiro
  * "o" Uma frutinha
  * "G" Um fantasma
* A primeira casa do tabuleiro sempre será um espaço vazio (".")
* 2 <= N <= 50

**Saída**:

Seu programa deve retornar um único inteiro, o número máximo de frutinhas que o Pacman pode levar para casa após a fase.

**Exemplo**:

Entrada:

[3, ".o.", "oGG", "ooo"]

Para melhor visualização do tabuleiro:  

    .o.
    oGG
    ooo

Saída:

4

### Explicação da Implementação da Função:


#### Funcionamento da Função

1. **Inicialização de Variáveis:**
   - `mochila`: Mantém o número atual de frutas coletadas.
   - `max_fruits`: Armazena o número máximo de frutas coletadas até o momento..

2. **Percorrendo as Linhas do Tabuleiro:**
   - Dependendo se a linha é par ou ímpar, Pacman percorre a linha da esquerda para a direita (LEFT -> RIGHT) ou da direita para a esquerda (RIGHT -> LEFT).

3. **Coletando Frutas e Encontrando Fantasmas:**
   - Se o pacman encontra uma fruta ('o'), ele a coleta, incrementando a variável `mochila`.
   - Se Pacman encontra um fantasma ('G'), ele perde todas as frutas coletadas até então, e a variável `mochila` é resetada para 0.

4. **Retorno do Resultado:**
   - Após percorrer todas as células do tabuleiro, a função retorna o valor de `max_fruits`.



```python
def pacman(arr) -> int:
  """
  Nesta logica o pacman não tem condição de parada, e retora o número maximo de frutas que é possível encontrar neste tabuleiro.
  """
  N = arr[0]
  tabuleiro = arr[1:]
  mochila = 0
  max_fruits = 0

  # percorrendo as linhas
  for i in range(N):
    #Se o indice da linha for par, ele percorre LEFT -> RIGHT
    if i % 2 == 0:
      start = 0
      end = N
      step = 1
    else: #RIGHT -> LEFT
      start = N - 1
      end = -1
      step = -1
    #percorrendo as colunas
    for j in range(start, end, step):
      next = tabuleiro[i][j]
      if next == 'o':
        mochila +=1
        if mochila > max_fruits:
          max_fruits = mochila
        #print("Ganhei uma fruta")
      elif next == 'G':
        mochila = 0
        #print("Perdi tudo")

  return max_fruits
```

### Explicação da Implementação da Função `pacman_update` com Condição de Parada

Se o Pacman tem a oportunidade de parar a qualquer momento do GAME então ele deve maximizar sua chance de conseguir carregar o máximo de frutas e levar em consideração as frutas que ele ja tem, pensando nisso criei uma nova função `pacman_update(arr, limite_proporcao=0.5)`, a diferença é que ele para de jogar quando o risco de perder todas as frutas é alto.

```python
def pacman_update(arr, limite_proporcao=0.5) -> int:
    """
    Neste caso, o jogo do PacMan tem uma condição de parada chamada Limite de proporção que é definido como:
    limite_proporção = frutas coletadas / celulas restantes

    Um limite de proporção baixo (por exemplo, 0.1 ou 0.2) significa que o Pacman está configurado para parar mais cedo.

    Um limite de proporção alto (por exemplo, 0.5 ou 0.8) significa que Pacman está configurado para continuar jogando,
    até que tenha coletado uma grande quantidade de frutas em relação ao número de células restantes.contrar um fantasma.
    """
    N = arr[0]
    tabuleiro = arr[1:]
    mochila = 0

    # Número total de células no tabuleiro
    total_celulas = N * N

    # Células percorridas até o momento
    celulas_percorridas = 0

    # percorrendo as linhas
    for i in range(N):
        if i % 2 == 0:
            start = 0
            end = N
            step = 1
        else:
            start = N - 1
            end = -1
            step = -1

        # percorrendo as colunas
        for j in range(start, end, step):
            celulas_percorridas += 1
            next = tabuleiro[i][j]
            if next == 'o':
                mochila += 1

            # Calcula a proporção de frutas coletadas
            celulas_restantes = total_celulas - celulas_percorridas
            proporcao = mochila / celulas_restantes if celulas_restantes > 0 else 1

            # Condição de parada baseada na proporção
            if proporcao >= limite_proporcao:
                return mochila

            elif next == 'G':
                mochila = 0

    return mochila
```

### Explicação Breve da Função `merge_sort`

A função `merge_sort(lista)` implementa o algoritmo de ordenação Merge Sort, que utiliza da técnica **divisão e conquista**. Essa técnica é uma abordagem fundamental para resolver problemas complexos ao dividi-los em subproblemas menores e mais fáceis.

#### Complexidade Geral do Merge Sort:

- **Complexidade Temporal:** O(n log n) - Em todas as situações (melhor, pior e caso médio), a complexidade do Merge Sort é O(n log n). Isso ocorre porque o algoritmo sempre divide a lista em duas metades (log n divisões) e, em seguida, realiza uma mesclagem de todos os elementos (n operações).
  
- **Complexidade Espacial:** O(n) - O algoritmo requer espaço adicional proporcional ao tamanho da lista, pois precisa armazenar as sublistas temporárias durante o processo de mesclagem.



```python
from typing import Any, List
```

```python
def merge_sort(lista):
    """
    Adiante, precisaremos conhecer previamente a complexidade do algoritmo de ordenação, porquê alguns algooritmos de busca requerem a lista previamente ordenada.
    Ordena uma lista usando o algoritmo Merge Sort, que é um exemplo clássico da técnica de divisão e conquista.
    Esta técnica divide o problema em problemas menores, resolve-os recursivamente e combina suas soluções para formar a solução do problema original.
    Complexidade Geral: O(n log n)
    Melhor caso: O(n log n)
    Pior caso: O(n log n)
    Espaço de Execução: O(n), devido ao espaço necessário para armazenar as sublistas temporárias durante a mesclagem.
    A complexidade é a mesma em todos os casos porque o algoritmo sempre divide a lista em duas metades e mescla-as, processando cada elemento várias vezes em diferentes níveis de recursão.

    """
    if len(lista) <= 1:  # O(1) - Verifica se a lista é mínima (base da recursão).
        return lista  # O(1) - Retorna a lista como está se contém 0 ou 1 elemento.

    meio = len(lista) // 2  # O(1) - Calcula o índice do meio da lista.
    esquerda = merge_sort(lista[:meio])  # T(n/2) - Recursivamente ordena a metade esquerda.
    direita = merge_sort(lista[meio:])  # T(n/2) - Recursivamente ordena a metade direita.

    return merge(esquerda, direita)  # O(n) - Mescla as duas metades ordenadas.

def merge(esquerda, direita):
    resultado = []  # O(1) - Cria uma lista vazia para os resultados.
    i = j = 0  # O(1) - Inicializa os contadores para as sublistas esquerda e direita.

    while i < len(esquerda) and j < len(direita):  # O(n) - Executa até que uma das sublistas esteja completamente mesclada.
        if esquerda[i] < direita[j]:  # O(1) - Compara elementos das duas sublistas.
            resultado.append(esquerda[i])  # O(1) - Adiciona o menor elemento ao resultado.
            i += 1  # O(1) - Incrementa o índice da sublista esquerda.
        else:
            resultado.append(direita[j])  # O(1) - Adiciona o menor elemento ao resultado.
            j += 1  # O(1) - Incrementa o índice da sublista direita.

    resultado.extend(esquerda[i:])  # O(k) - Adiciona os elementos restantes da esquerda ao resultado.
    resultado.extend(direita[j:])

    return resultado  # O(1) - Retorna a lista mesclada e ordenada.
```

### Explicação Breve da Função `busca_linear`

A função `busca_linear(lista, elemento)` implementa o algoritmo de busca linear, que é um dos métodos mais simples de busca em uma lista. A busca linear percorre a lista elemento por elemento até encontrar o elemento desejado ou até que todos os elementos tenham sido verificados.

#### Como Funciona:

1. **Percorrendo a Lista:**
   - A função utiliza um laço `for` para iterar sobre cada elemento da lista. Para cada elemento, ela verifica se corresponde ao `elemento` que está sendo procurado.

2. **Comparação e Retorno:**
   - Se o elemento é encontrado, a função retorna imediatamente o índice desse elemento na lista.
   - Se o elemento não for encontrado após a verificação de todos os elementos, a função retorna `-1`, indicando que o elemento não está presente na lista.

#### Considerações sobre Tipos de Dados:

- **Inteiros**
- **Float**  
- **Strings**

```python
def busca_linear(lista, elemento) -> int:
    """
    Melhor caso: O(1) - Quando o elemento está no início da lista.
    Pior caso: O(n) - Quando o elemento não está na lista ou está no final.
    Espaço de execução: O(1) Ela percorre a lista elemento por elemento sem necessitar de estruturas de dados adicionais ou recursividade.
    A lista não precisa estar ordenada.

    A função funciona tanto com tipos de dados inteiros quanto com strings. Para inteiros, a comparação é direta e baseada em valores.
    Para strings, a comparação é feita utilizando a tabela ASCII para avaliar as iniciais de cada palavra.
    """
    # Laço que percorre cada elemento da lista.
    for indice, valor in enumerate(lista):  # O(n) no pior caso, pois pode percorrer toda a lista
        if valor == elemento:               # O(1), comparação simples entre dois valores
            return indice                   # O(1), retorno do índice encontrado

    return -1                               # O(1), retorno de -1 se o elemento não for encontrado
```

### Explicação Breve da Função `busca_ordenada`

A função `busca_ordenada(lista, elemento)` combina ordenação e busca binária para encontrar um elemento em uma lista de forma eficiente.

O conceito da busca binária é sempre dividir na metade os dados e verificar se ele está no meio, a importância de estar ordenado é que se o elemento não estiver no meio o algoritmo não vai saber em qual lado ele está, se é menor ou maior.

#### Complexidade:

- **Ordenação:** O(n log n)
- **Busca Binária:** O(log n)
- **Total:** O(n log n)




```python
def busca_ordenada(lista, elemento):
    """
    Melhor caso: O(1) - Quando o elemento está no meio da lista.
    Pior caso: O(log n) - Quando o elemento não está na lista ou está nas extremidades.
    Espaço de execução: O(1)A busca binária iterativa só precisa de algumas variáveis para armazenar os índices dos limites da busca (início, meio e fim).

    A complexidade da busca binária pode variar de O(nlogn) devido ao algoritmo utilizado para ordenação.
    Além disso, a busca binária pode ser implementada de forma recursiva, desde que
    a lista seja ordenada previamente.


    Assim como na função de busca linear as comparações com string são feitas da mesma forma.
    """
    # Ordena a lista usando o algoritmo de ordenação Merge Sort
    lista = merge_sort(lista)  # O(n log n) - Esta linha ordena a lista.

    # Iniciando os ponteiros para início e fim da lista
    esquerda = 0  # O(1) - Atribuição inicial do ponteiro esquerdo.
    direita = len(lista) - 1  # O(1) - Atribuição inicial do ponteiro direito.

    while esquerda <= direita:  # O(log n) - Este laço ocorre log n vezes no máximo.
        meio = (esquerda + direita) // 2  # O(1) - Calcula o índice médio.
        if lista[meio] == elemento:  # O(1) - Verifica se o elemento está no índice médio.
            return meio  # O(1) - Retorna o índice do elemento encontrado.

        elif lista[meio] < elemento:  # O(1) - Compara o elemento com o valor no meio.
            esquerda = meio + 1  # O(1) - Ajusta o ponteiro esquerdo para reduzir a metade da busca.

        else:  # O(1) - Caso o valor do meio seja maior que o elemento.
            direita = meio - 1  # O(1) - Ajusta o ponteiro direito para reduzir a metade da busca.

    return -1  # O(1) - Retorna -1 se o elemento não for encontrado.
```

### Explicação Breve da Função `busca_saltos`

A função `busca_saltos(lista, elemento, passo)` implementa o algoritmo de busca por saltos, uma técnica eficiente para encontrar um elemento em listas ordenadas.

#### Como Funciona:

1. **Ordenação:**
   - A lista é ordenada usando o Merge Sort para garantir que a busca por saltos funcione corretamente.

2. **Busca por Saltos:**
   - A função percorre a lista em blocos de tamanho `passo`
   
3. **Ajuste de Passo:**
   - **Passo Grande:** Menos saltos, mas a busca linear pode ser mais longa.
   - **Passo Pequeno:** Mais saltos, mas a busca linear é mais curta.

#### Complexidade:

- **Melhor Caso:** O(1) - Se o elemento está no primeiro bloco.
- **Pior Caso:** O(√n) - Se o elemento está no último bloco ou não está na lista.
- **Espaço de Execução:** O(1) - Apenas algumas variáveis são usadas para o controle dos índices.

```python
def busca_saltos(lista, elemento, passo):
    """
      Melhor caso: O(1) - Quando o elemento está no primeiro bloco.
      Pior caso: O(√n) - Quando o elemento não está na lista ou está nos últimos blocos.
      Espaço de execução: O(1) Semelhante à busca linear, ela utiliza um número limitado de variáveis para controlar o índice e os limites de busca, mantendo assim um espaço constante.

      Passo Grande -> Com um passo grande, o número de saltos necessários para percorrer a lista é menor.
      No entanto, se o elemento não é encontrado durante os saltos, a busca linear no último bloco pode ser longa, pois o bloco será grande.

      Passo Pequeno -> Com um passo pequeno, a lista é dividida em mais blocos, aumentando o número de saltos necessários para percorrer a lista.
      Por outro lado, se o elemento não é encontrado durante os saltos, a busca linear que se segue cobrirá um bloco menor,
      o que pode ser mais rápido se o elemento está próximo do início do bloco.

      Como nas duas funções anteriores, as strings são comparadas com a iniciais e os valores da tabela ASCII.
      """
    lista = merge_sort(lista)  # O(n log n) - Ordena a lista usando o algoritmo de ordenação Merge Sort.
    n = len(lista) # O(n) - Obtém o tamanho da lista, operação linear.
    indice_atual = 0 # O(1) - Inicia o índice atual para a busca.
    indice_anterior = 0  # O(1) - Inicializa o índice anterior para referência na busca linear subsequente.


    # Procura um bloco onde o número pode estar.
    while indice_atual < n and lista[indice_atual] < elemento:
        indice_anterior = indice_atual  # O(sqrt(n)) - Percorre os blocos com tamanho definido por 'passo'.
        indice_atual += passo # O(1) - Incrementa o índice atual pelo tamanho do passo.
        # Se o índice atual ultrapassar a lista, ajusta para o último índice válido.
        if indice_atual >= n: # O(1) - Verifica se o índice ultrapassou o tamanho da lista.
            indice_atual = n - 1 # O(1) - Ajusta o índice para o último elemento da lista.
            break

    # Busca linear no bloco encontrado.
    for i in range(indice_anterior, min(indice_atual + 1, n)): # O(passo) - Percorre o bloco onde o elemento pode estar.
        if lista[i] == elemento: # O(1) - Verifica se o elemento atual é o elemento buscado.
            return i # O(1) - Retorna o índice do elemento encontrado.

    return -1 # O(1) - Retorna -1 se o elemento não for encontrado.
```

### Explicação Breve da Função `busca_hibrida`

A função `busca_hibrida(lista, elemento, passo)` combina duas técnicas de busca eficientes—**busca por saltos** e **busca binária**—para localizar um elemento em uma lista ordenada.

#### Como Funciona:

1. **Busca por Saltos:**
   - A função divide a lista em blocos de tamanho `passo`.
   - Percorre os blocos, verificando se o elemento pode estar dentro de um deles. Assim que o bloco provável é encontrado, a busca por saltos termina.

2. **Busca Binária no Bloco:**
   - Dentro do bloco identificado, a função realiza uma busca binária para encontrar a posição exata do elemento, aproveitando a eficiência da busca binária em listas ordenadas.

#### Complexidade:

- **Melhor Caso:** O(1) - Quando o elemento está no primeiro bloco.
- **Pior Caso:** O(√n) - Se o elemento está no último bloco ou não está na lista.
- **Espaço de Execução:** O(1) - A função utiliza apenas algumas variáveis adicionais, mantendo um uso constante de espaço.


```python
def busca_hibrida(lista, elemento, passo) -> Any:
    """
    Melhor caso: O(1) - Quando o elemento está no primeiro bloco.
    Pior caso: O(√n) - Quando o elemento não está na lista ou está nos últimos blocos.
    Espaço de execução: O(1), pois é a junção de dois algoritmos que possuem complexidade espacial O(1).

    Combina a busca por saltos e a busca binária para encontrar o índice de um elemento em uma lista ordenada.
    A lista é primeiro ordenada usando o Merge Sort, depois passa por uma busca por saltos para encontrar um bloco
    e no final uma busca binária é aplicada dentro desse bloco.
    """

    # Ordena a lista usando Merge Sort
    lista = merge_sort(lista)  # O(n log n) - Ordena a lista, complexidade dominante para grandes n.

    # Busca por saltos para determinar o bloco aproximado
    n = len(lista)  # O(1) - Obtém o tamanho da lista.
    bloco_inicio = 0  # O(1) - Inicializa o índice de início do bloco.

    while bloco_inicio < n and lista[bloco_inicio] < elemento:
        bloco_fim = min(bloco_inicio + passo, n - 1)  # O(1) - Calcula o fim do bloco, garantindo que não ultrapasse o tamanho da lista.
        if lista[bloco_fim] >= elemento:
            break  # O(1) - Sai do loop se o elemento está dentro do bloco.
        bloco_inicio = bloco_fim + 1  # O(1) - Atualiza o início do próximo bloco.

    if bloco_inicio >= n or lista[bloco_inicio] > elemento:
        return -1  # O(1) - Retorna -1 se o elemento não está no bloco ou ultrapassa o limite.

    # Busca binária dentro do bloco identificado
    esquerda, direita = bloco_inicio, bloco_fim  # O(1) - Define os ponteiros para a busca binária.
    while esquerda <= direita:  # O(log(passo)) - A busca binária ocorre dentro do intervalo definido pelo passo.
        meio = (esquerda + direita) // 2  # O(1) - Encontra o meio do intervalo.
        if lista[meio] < elemento:
            esquerda = meio + 1  # O(1) - Ajusta o ponteiro esquerdo para o subintervalo direito.
        elif lista[meio] > elemento:
            direita = meio - 1  # O(1) - Ajusta o ponteiro direito para o subintervalo esquerdo.
        else:
            return meio  # O(1) - Retorna o índice do elemento encontrado.

    return -1  # O(1) - Retorna -1 se o elemento não for encontrado após a busca binária.
```

# Questão 3: Autômato de Pilha

Tarefa:

Implemente um autômato de pilha que aceite a linguagem { $L = a^i b^j c^k
 : i ≠ 0, i < j, k < j$ } sobre o alfabeto { $a, b, c$ }. O autômato deve seguir as seguintes regras:

Descrição:

O autômato de pilha é uma máquina teórica que usa uma pilha para ajudar a decidir se uma palavra pertence à linguagem. Esse tipo de autômato é capaz de reconhecer linguagens que não podem ser reconhecidas por autômatos finitos simples, como linguagens com padrões aninhados e outras formas complexas de dependência entre símbolos.

Regras
O autômato deve aceitar palavras que começam com pelo menos um símbolo 'a' ($i \neq 0$).
O número de símbolos 'b' deve ser maior que o número de símbolos 'a' ($i < j$).
O número de símbolos 'b' deve ser maior que o número de símbolos 'c' ($k < j$).
O autômato deve ler a palavra de entrada, caracter por caracter, e utilizar a pilha para controlar a contagem e a ordem dos símbolos lidos.

Requisitos:

* Alfabeto: O autômato de pilha deve trabalhar com o alfabeto { $a$, $b$, $c$ }.
* Linguagem: A linguagem que o autômato deve aceitar é definida por { $L$ = $a^i$ $b^j$ $c^k$ : $i \neq 0$, $i < j$, $k < j$ }.
* Condições da Linguagem:
  1. A palavra deve começar com pelo menos um símbolo 'a' ($i \neq 0$).
  2. O número de símbolos 'b' deve ser maior que o número de símbolos 'a' ($i < j$).
  3. O número de símbolos 'b' deve ser maior que o número de símbolos 'c' ($k < j$).
* Comportamento do Autômato:
  1. O autômato deve usar uma pilha para rastrear e validar a contagem dos símbolos conforme as regras da linguagem.
  2. Ele deve ser capaz de aceitar ou rejeitar uma palavra após processar todos os símbolos da entrada.

# Exemplos
* Entrada 1
  * Entrada: "aabbbc"
  * Saída: "Sucesso!"

* Entrada 2
  * Entrada: "aabbcc"
  * Saída: "Palavra nao aceita"

## Automato de Pilha:

- Objetivo é fazer com que ele aceite uma linguagem que a regra seja a quantidade de a's ser menor que a quantidade de b's e a quantidade de c's ser menor que a quantidade de b's.
- Iniciando com o caractere "a", o automato vai empilhar o "a" caso a pilha esteja no estado inicial ou com "$" no topo da pilha.
- Quando ele terminar de ler a quantidade de "a" ele vai partir para a leitura dos "b" quando ele ler um "b" ele deve desempilhar um "a", quando sobrar "b" na pilha isso indica que a quantidade de "b" é maior do que a quantidade de "a".
- Mesma coisa para os "c".

```python
class Automatopilha:
    def __init__(self):
        self.stack = ['$']

    def execute(self, input_string):
        for char in input_string:
            #print(char)

            if char == 'a':
              # To lendo A, tem $ no topo da pilha, empilha A
              if self.stack[-1] == '$':
                self.stack.append('a')
              # To lendo A, tem A no topo da pilha, empilha A
              elif self.stack[-1] == 'a':
                self.stack.append('a')


            elif char == 'c':
                #To lendo C, tem B no topo da pilha, desempilha B
                if self.stack[-1] == 'b':
                    self.stack.pop()
                # To lendo C, tem $ no topo da pilha, empilha C
                elif self.stack[-1] == '$':
                    self.stack.append('c')


            elif char == 'b':
                # To lendo B, tem A no topo da pilha, desempilha A
                if self.stack[-1] == 'a':
                    self.stack.pop()

                # To lendo B, tem $ no topo da pilha, empilha B
                elif self.stack[-1] == '$':
                    self.stack.append('b')

                # To lendo B, tem B no topo da pilha, empilha B
                elif self.stack[-1] == 'b':
                    self.stack.append('b')

                elif self.stack[-1] == 'c':
                    self.stack.pop()
            print(self.stack)


        # Aceita se o topo da pilha é '$' (todo o resto foi desempilhado corretamente)
        if len(self.stack) == 1 and self.stack[-1] == '$':
            return "Aceita"
        else:
            return "Rejeitada"
```

```python
from collections import Counter

def calcular_moda_temperamentos(data, column_name):
    # Separar os temperamentos e contar a frequência de cada um
    temperamentos = []
    for item in data[column_name].dropna():
        temperamentos.extend(item.split(', '))
    #visualizar todos os temperamentos
    #print(temperamentos)

    # Utilizando Counter para contar a frequência
    temperamento_count = Counter(temperamentos)
    print(temperamento_count)

    # Encontrando o temperamento mais comum
    moda = temperamento_count.most_common(1)

    print(f"O temperamento mais comum é: {moda[0][0]} com {moda[0][1]} ocorrências.")
    return moda
```

## Medidas de Dispersão


Três medidas fundamentais de dispersão mais comuns:

- 1. **Amplitude**
  - A amplitude é a diferença entre o maior e o menor valor em um conjunto de dados. Ela fornece uma medida rápida do intervalo de valores.
  - **Fórmula da Amplitude:**
  $$ \text{Amplitude} = \max(x) - \min(x) $$


- 2. **Variância**

  - A variância mede a média dos quadrados das diferenças entre cada valor e a média do conjunto de dados. Ela fornece uma noção de quão dispersos os dados estão em torno da média.

  - **Fórmula da Variância \( $\sigma^2 $):**
  $$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 $$
  onde \( $\mu $) é a média dos dados, e \( N \) é o número total de observações.

- 3. Desvio Padrão

  - O desvio padrão é a raiz quadrada da variância e é usado para medir a dispersão dos dados da mesma forma que a variância, mas na mesma unidade dos dados.

  - **Fórmula do Desvio Padrão \( $\sigma $):**
  $$ \sigma = \sqrt{\sigma^2} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2} $$

```python
def calcular_medidas_dispersao(data,colunas):
    for coluna in colunas:
        amplitude = data[coluna].max() - data[coluna].min()
        variancia = data[coluna].var()
        desvio_padrao = data[coluna].std()
        print(f"Para a coluna '{coluna}':")
        print(f"  Amplitude: {amplitude:.4f}")
        print(f"  Variância: {variancia:.4f}")
        print(f"  Desvio Padrão: {desvio_padrao:.4f}\n")
```

```python
calcular_medidas_dispersao(dataset,colunas_numericas)
```

- Agora iniciarei o tratamento dos outliers, desde a identificação e utilizando a técnica de Capping

```python
#utilizar técnica de Capping nas colunas numéricas que possuem outliers
def capping(dataset,column):
    """
    Capping consiste em substituir o valor de um outlier por valores que estão no limite inferior e superior
    Desde modo, não perco dados e evito que os outleirs influenciem negativamente.
    """
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    dataset[column] = dataset[column].apply(lambda x: upper_bound if x > upper_bound else lower_bound if x < lower_bound else x)
    return dataset
```

```python
#Utilizando em min_weight e max_weight
dataset = capping(dataset,'min_weight')
dataset = capping(dataset,'max_weight')

# utilziando em min_expectancy e max_expectancy
dataset = capping(dataset,'min_expectancy')
dataset = capping(dataset,'max_expectancy')

#boxplot após capping
plt.figure(figsize=(12, 8))
dataset[numerical_columns].boxplot()
plt.title('Boxplot of Numerical Features')
plt.ylabel('Values')
plt.grid(False)
plt.show()
```

5.1.4 - Lematização

#### Lematização é o processo de reduzir uma palavra para sua forma base ou lema. Por exemplo, a palavra "correria" seria lematizada para "correr", enquanto "melhor" seria lematizado para "bom".



```python
# Importando os módulos necessários do NLTK
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('popular')
```

```python
# Função para mapear a tag de parte da fala do NLTK para a tag usada pelo WordNet lemmatizer
def get_wordnet_pos(word):
    """
    Função para mapear a tag de parte da fala do NLTK para a forma aceita pelo WordNet lemmatizer.
    Ela serve para identificar o que são substantivos, o que são verbos,...
    """
    # Utiliza o POS tagger do NLTK para obter a tag da palavra, que é o segundo elemento do par retornando pelo pos_tag
    tag = nltk.pos_tag([word])[0][1][0].upper()
    # Dicionário para mapear tags do NLTK para tags do WordNet
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    # Retorna a tag do WordNet, ou NOUN como padrão se a tag não estiver no dicionário
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
# Função para lematizar
def lemmatize_tokens(tokens):
    """
    Função para lematizar uma lista de tokens.
    """

    # Inicializando o objeto WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()
    # Lematizar cada token usando a função lemmatize do lematizador, especificando a tag de parte da fala
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]
    return lemmatized_tokens
```

```python
def get_most_common_words(df, column_tokens, target):
    """
    Função para encontrar as palavras mais comuns em uma lista de listas de tokens,
    filtradas por uma classificação específica.
    """
    # Filtrar o DataFrame para incluir apenas as linhas com a classificação desejada
    filtered_df = df[df['classification'] == target]

    all_words = []
    for tokens in filtered_df[column_tokens]:
        # Agregando todos os tokens da classificação específica
        all_words.extend(tokens)

    return Counter(all_words).most_common(10)
```

- Mesmo após utilizar a escala logarítmica pode-se observar a persistência de alguns outliers mesmo após a tentativa de suavização, por isso, vamos utilizar a técnica de CAPPING novamente.

```python
#utilizar técnica de Capping nas colunas numéricas que possuem outliers
def capping(dataset,column):
    """
    Capping consiste em substituir o valor de um outlier por valores que estão no limite inferior e superior
    Desde modo, não perco dados e evito que os outleirs influenciem negativamente.
    """
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    dataset[column] = dataset[column].apply(lambda x: upper_bound if x > upper_bound else lower_bound if x < lower_bound else x)
    return dataset
```

```python
#Utilizando em revenue
movies = capping(movies,'log_revenue')
movies = capping(movies,'log_budget')
movies = capping(movies,'runtime')
movies = capping(movies,'vote_count')


#boxplot após capping
plt.figure(figsize=(10, 8))
for i, column in enumerate(check_outliers, 1):
    plt.subplot(2, 2, i)  # Subplots em uma grade 2x2
    sns.boxplot(y=movies[column])
    plt.title(f'Box Plot de {column}')
    plt.ylabel('Valores')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```

## Regressão Linear Múltipla

### Introdução

A **Regressão Linear Múltipla** é uma extensão da regressão linear simples que permite prever uma variável dependente (`Y`) com base em duas ou mais variáveis independentes (`X1, X2, ..., Xn`). Esse modelo é fundamental em estatística e aprendizado de máquina para entender a relação entre a variável alvo e várias variáveis preditoras.

### Problema de Regressão

O problema de regressão envolve a previsão de um valor contínuo, no caso: `log_revenue`. Por exemplo, prever a receita de um filme com base em características como orçamento, popularidade e gênero.

- Escolha empírica da quantidade de `k-clusters`, existem alguns métodos para encontrar um `K` ideal, como `método do joelho`, como também existem métodos que aproximam a quantidade de neurônios em uma rede neural. Porém, a quantidade estimada, dependendo do método é um limite superior, resultando em um modelo desnecessariamente mais robusto.
